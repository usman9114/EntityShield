# Loading UEBA anomilies

```python
anomaly_score = pd.read_csv('final_anomaly_score.csv')
```

```python
def preprocess_http(df):
    df = df.copy()
    df['phishing_attack'] = df['phish'].apply(lambda x :False if ('Not Phishing'== x[0].split(':')[2].replace('}','').strip()) else True )*1
    df['DGA'] =df['dga_http'].apply(lambda x :True if(x=='unsafe') else False )*1
    df['Bot_agent'] =df['bot'].apply(lambda x :x[0] *1)
    df =df[['id.orig_h','id.resp_h','phishing_attack','DGA','Bot_agent']]
    return df
```

```python
preprocess_http(http)
```

# Preprocessing / filtering only threats

```python
def preprocessing(df,df2=pd.DataFrame()):
    df = df.copy().reset_index(drop=True)
#     df2 = df2.copy().reset_index(drop=True)

    
    df = df[df[['phishing_attack','DGA','Bot_agent']].values.sum(axis=1) != 0] # 3X faster 
    try:
        df2['DGA'] =df2['dga_dns'].apply(lambda x :True if(x=='unsafe') else False )*1
        df2 = df2[df2['DGA'].values!=0]
        df2 =df2[['id.orig_h','id.resp_h','DGA']]

    except:
        pass
    df = df[['id.orig_h','id.resp_h','phishing_attack','DGA','Bot_agent']]

    temp= pd.concat([df,df2],axis=0,sort=True)
    return temp[['id.orig_h','id.resp_h','phishing_attack','DGA','Bot_agent']].fillna(0)
    
```

```python
score_set= preprocessing(http)
```

```python
final_score = pd.concat([score_set,anomaly_score],sort=True)
```

```python
final_score.fillna(0,inplace=True)
```

### finding private IP, as score will only be calculated for local IP address

```python
ip_list = list(set([f for f in final_score['id.orig_h'] if ipaddress.ip_address(f).is_private]\
+[f for f in final_score['id.resp_h'] if ipaddress.ip_address(f).is_private]))
```

```python
ip_list
```

### Subseting dataset and calculating score for each IP, either in source or destionation

```python
temp_df_score=pd.DataFrame()
for ip in ip_list:
    temp = final_score.loc[(final_score['id.orig_h']==ip)|(final_score['id.resp_h']==ip)]
    temp_score = temp.groupby(temp['id.orig_h']).sum()
    temp_df_score = pd.concat([temp_score,temp_df_score],axis=0)
temp_df_score =temp_df_score.reset_index()   
temp_df_score.columns = ['IP Address','Bot_agent','DGA','conn_anomaly','dns_anomaly','http_anomaly','phishing_attack']
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
        #return '{p:.2f}%'.format(p=pct)

    return my_autopct
```

```python
# imports
%matplotlib inline

from ipywidgets import interactive
import pandas as pd
import numpy as np
# from jupyterthemes import jtplot

# Sample data
np.random.seed(123)


#jtplot.style()

import ipywidgets as widgets
from IPython.display import display

def multiplot(ip):
    row = temp_df_score.loc[temp_df_score['IP Address']==ip]
    plt.figure(figsize=(8,8))
    explode = (0.05, 0.05, 0.05, 0.05,0.05,0.05)
    size = list(row.values[0][1:])
    label= list(row.columns[1:])
    title = plt.title('Risk scoring for IP: {}'.format(ip),fontdict={'fontsize':25})
    title.set_ha("left")
    plt.gca().axis("equal")
    pie = plt.pie(size,autopct=make_autopct(size), startangle=0,textprops={'fontsize': 14},explode=explode)
    labels=label
    plt.legend(pie[0],labels, bbox_to_anchor=(1,0.5), loc="center right", fontsize=15, 
               bbox_transform=plt.gcf().transFigure)
    plt.subplots_adjust(left=0.0, bottom=0.1, right=.76)

interactive_plot = interactive(multiplot, ip=ip_list)
output = interactive_plot.children[-1]
output.layout.height = '500px'
interactive_plot
```

```python
def weigth_scoring(df,**args):
    df=df.copy()
    df['Bot_agent'] = df['Bot_agent']*args['Bot_agent']
    df['DGA'] = df['DGA']*args['DGA']
    df['conn_anomaly'] = df['conn_anomaly']*args['conn_anomaly']
    df['dns_anomaly'] = df['dns_anomaly']*args['dns_anomaly']
    df['http_anomaly'] = df['http_anomaly']*args['http_anomaly']
    df['phishing_attack'] = df['phishing_attack']*args['phishing_attack']
    return df
```

```python
final_score = weigth_scoring(temp_df_score,Bot_agent=0.01,DGA=4,conn_anomaly=3.5,dns_anomaly=2,
              http_anomaly=2,phishing_attack=1)
final_score['total'] = final_score[final_score.columns[1:]].sum(axis=1)
```

```python
final_score = final_score.sort_values(by='total',ascending=False)
```
