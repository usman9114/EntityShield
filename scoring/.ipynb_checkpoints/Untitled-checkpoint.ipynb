```python
import pandas as pd
import os
```

```python
df_conn_score = pd.DataFrame()
for i in os.listdir('conn'):
    temp = pd.read_csv('conn/'+i)
    df_conn_score = pd.concat([temp,df_conn_score],axis=0)
```

```python
df_http_score = pd.DataFrame()
for i in os.listdir('http'):
    temp = pd.read_csv('http/'+i)
    df_http_score = pd.concat([temp,df_http_score],axis=0)
```

```python
df_dns_score = pd.DataFrame()
for i in os.listdir('dns'):
    temp = pd.read_csv('dns/'+i)
    df_dns_score = pd.concat([temp,df_dns_score],axis=0)
```
